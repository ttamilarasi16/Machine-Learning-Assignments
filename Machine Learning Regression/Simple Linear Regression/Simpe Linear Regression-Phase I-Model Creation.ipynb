```python
#pandas is a python library, importing that helps to work with/handle data easily and effectively from files/tables.

import pandas
```

```python
#here changing the name pandas as pd, just to call easily.

import pandas as pd
```

```python
#DATA_COLLECTION
#assigning the data from the csv file to the variable "dataset" using read_csv function.

dataset = pd.read_csv("Salary_Data.csv")
```

```python
#SPLIT INPUT & OUTPUT
#here assigning the input and output for separate variables inorder to spliting them.

independent_variable = dataset[["YearsExperience"]]
dependent_variable = dataset[["Salary"]]
```

```python
#SPLIT TRAIN & TEST
#sklearn is a python library, model_selection is a class from that and train_test_split is a function present inside the class, 
#"from" is a keyword that is used to import a specific function inorder to reduce the memory space by not importing all the functions from a library.
#Next is to split the training and test set and store it into 4 different variables (x_train, y_train, x_test, y_test)
#the test size is 0.3 which is 30%, so obviously the training data will be the size of 0.70 that is 70%

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(independent_variable, dependent_variable, test_size = 0.30, random_state = 0)
```

```python
#TRAIN THE MODEL
#LinearRegression is a function present inside the sklearn library
#here assigning LinearRegression function to a variable called SLR model
#And training the model using the function called fit by giving the training data as parameters

from sklearn.linear_model import LinearRegression
SLR_Model = LinearRegression()
SLR_Model.fit(x_train,y_train)
```

```python
#to find the weight

weight = SLR_Model.coef_
weight
```

```python
#to find the bias

bias = SLR_Model.intercept_
bias
```

```python
#EVALUATE METRICS

y_pred = SLR_Model.predict(x_train)

#calculating r2 value

from sklearn.metrics import r2_score
r2_score_value = r2_score(y_train, y_pred)
r2_score_value
```

```python
#SAVE MODEL
#importing pickle library and create the file

import pickle
model_filename = "Finalized_model_SLR.sav"

#dumping the model into the file

pickle.dump(SLR_Model,open(model_filename, 'wb'))
```
